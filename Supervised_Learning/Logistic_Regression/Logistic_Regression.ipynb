# Logistic Regression

---

Logistic Regression is commonly used to predict that an instance belongs to a particular class (e.g., what is the probability that this iris is a virginica?). If the model predicts the probability is greater than 50%, then the model predicts that the instance belongs to that class, otherwise it does not. It uses a 1 if it does, and a 0 if not, making this a **binary classifier**.

---

## How Does it Work?

---



Like a Linear Regression Model, a Logistic Regression model calculates a weighted sum of the input features (plus a bias term). Instead of outputting the result of this computation like a Linear Regression model, it instead outputs the **logistic** of this result. The equation below shows the Logistic Regression model estimate probability in vectorized form.

$\hat{p} = h_{\theta}(x) = \sigma(x^{T}\theta) $

---

The logistic, denoted as $\sigma$, is a *sigmoid function* that outputs a number between 0 and 1. The Logistic function is defined below.


$\sigma(t) = \frac{1}{1 + exp (-t)} $

---

Once the Logistic Regression model has estimated $\hat{p} = h_{\theta}(x)$, the probability that an instance **x** belongs to the positive class, it can easily make it's prediction, $\hat{y}$ easily.

\begin{equation}
  \hat{y}=\left\{
  \begin{array}{@{}ll@{}}
    0, & \text{if}\ \hat{p} \leq 0.5 \\
    1, & \text{if}\ \hat{p} \geq 0.5
  \end{array}\right.
\end{equation} 

---

## Training and Cost Function

---

We now understand how the model estimates probabilities and makes predictions, but how does it train? Our objective of training is to set the parameter $\theta$ such that the model estimates **high** probabilities for positive instances (y=1) and low probabilities for negative instances (y=0). This idea is captured in the cost function for Logistic Regression as shown below:

\begin{equation}
  \text{c}(\theta)=\left\{
  \begin{array}{@{}ll@{}}
    -\text{log}(\hat{p}), & \text{if}\ y = 1 \\
    -\text{log}(1 - \hat{p}), & \text{if}\ y = 0
  \end{array}\right.
\end{equation} 

This cost function makes sense because -log(*t*) grows very large when *t* approaches 0, so the cost will be large (e.g. too much) if the model estimates a probability close to 0 when it should be 1, and vice versa. However, -log(*t*) is close to 0 when *t* is close to1, so the cost will be close to 0 if the estimated probability is close to 0 for a negative instance, and vice versa. This is exactly what we want for our model.


---

The cost function over the whole training set is the average cost over all training instances. It can be written in a single expression, called the *log loss*, as shown below:


$ J(\theta) = - \frac{1}{m}\sum_{i=1}^{m} [y^{(i)}log(\hat{p}^{(i)}) + (1 - y^{(i)})log(1 - \hat{p}^{(i)})] $

---

The cost function is convex, so **Gradient Descent** is guaranteed to find the global minimum. The partial derivatives of the cost function with regard to the $j^{th}$ model parameter $\theta_{j}$ are below:

$\frac{\partial}{\partial\theta_{j}}J(\theta) = \frac{1}{m}\sum_{i=1}^{m} (\sigma(\theta^{T}x^{(i)}) - y^{(i)})x_{j}^{(i)}  $

For each instance this equation computes the prediction error and multiplies it by the $j^{th}$ feature value, and then it computes the average over all training distances. Now that we have the gradient vector containing all partial derivatives, we can use it in the **Batch Gradient Descent** algorithm. 

---

## Executing the Model on the Iris Dataset

That's it! We now know how to train our Logistic Regression Model. For the purposes of our model, we'll employ the ```LogisticRegression``` module from scikit-learn. Documentation for this can be found at [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html). 

We're going to use the famous Iris Dataset to illustrate Logistic Regression. We'll import ```sklearn.datasets.load_iris()``` to load our Iris dataset, documentation for which can be found at [load_iris()](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris).

---

---

Let's begin by importing the necessary modules. We're going to build a classifier to detect the *Iris setosa* type based only on the ```petal length``` feature.

---


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline

iris = datasets.load_iris()


X = iris["data"][:, 2:3]
y = (iris["target"] == 0).astype(np.int)
````

---

We will now train a Logistic Regression model. We'll begin by importing ```sklearn.linear_model.LogisticRegression```. Documentation is linked above. We'll set our random_state to ensure reproducibility.

---

````python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state = 42)
log_reg.fit(X, y)
````

---

We'll now look at the model's estimated probabilities for flowers with petal lengths varying from 0 cm to 7 cm. We'll then plot a graph to visualize.

---

````python
X_new = np.linspace(0, 7, 1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_prob[:, 1] >= 0.5][0]

plt.figure(figsize = (12, 6))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_prob[:, 1], "g-", linewidth=2, label="Iris setosa")
plt.plot(X_new, y_prob[:, 0], "b--", linewidth=2, label="Not Iris setosa")
plt.xlabel("Petal length (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 7, -0.02, 1.02])
plt.show()
````

---

As we can see above, there is no overlap in probabilities for petal length in our model. This means that our Logistic Regression model shouldn't have too many issues predicting an *Iris setosa* based on petal length. It is highly confident (and outputs a high probability) that the flower is an *Iris setosa* for petal lengths at or less than 2 cm.

Let's use the ```predict``` module to see what our model outputs for petal lengths of 1.2 cm and 3.2 cm.

---

````python
log_reg.predict([[1.2], [3.2]])
````

---

For 1.2 cm, our model predicted 1 (e.g., the positive class) for *Iris setosa* and for 3.2 cm it predicted 0 implying that it is not an *Iris setosa*.
