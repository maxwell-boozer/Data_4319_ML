# K Means

---

KMeans is an Unsupervised Learning algorithm that is used when we have unlabeled data. The goal of KMeans is to find groups in the data, with the number of groups represented by **K** (chosen by the user initially). The algorithm then iteratively assigns each data point to one of K groups based on the features provided. The data points are *clustered* based on feature similarity. 



---

## Implementing KMeans

---

For the purposes of our example, we will be implementing the KMeans algorithm on a dataset containing Mall Customer data. We'll focus on the ```Annual_Income``` and ```Spending_Score``` attributes for each instance.

---

Let's begin by importing the necessary libraries.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

---

To determine the right k-value for our model we'll use the Within Cluster Sum of Squares (WCSS), which is a measure of the variability of the observations within each cluster. The equation for WCSS can be seen below:




We want to find the minimum number of clusters, so we'll use the Elbow Method to make our choice. 

---

To start we need to import the KMeans module from sklearn (```sklearn.cluster.KMeans```) documentation for which can be found at [KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html).

---

````python
from sklearn.cluster import KMeans
````

---

Next, we'll setup the elbow method to find the optimal K value. We'll also initialize with the ```init = 'k-means++'``` switch when we call KMeans to speed up convergence.

We'll also subset our data to focus on just ```Annual_Income``` and ```Spending_Score``` to visualize the KMeans model.

---

````python
X = mall.iloc[:, [3,4]].values

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, 
                    init = 'k-means++',
                    random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
````

---

Let's visualize our plot and see if we can spot the optimal K-value from the graph

---

````python
plt.figure(figsize=(12, 8))
plt.plot(range(1, 11), wcss)
````

---

We can see that after 5 clusters we do not observe a significant decrease in the WCSS plot

Let's now fit our KMeans model with 5 clusters and visualize the data. We'll also look at a plot where we've fit with a cluster at the far tail end of the plot and observe any difference.

---

````python
kmeans = KMeans(n_clusters = 5, 
                init = 'k-means++',
                random_state = 20)
pred = kmeans.fit_predict(X)
````

````python
plt.figure(figsize=(12, 8))
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 100, c = 'darkred', label = 'Cluster 1')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'navy', label = 'Cluster 2')
plt.scatter(X[pred == 2, 0], X[pred == 2, 1], s = 100, c = 'forestgreen', label = 'Cluster 3')
plt.scatter(X[pred == 3, 0], X[pred == 3, 1], s = 100, c = 'darkcyan', label = 'Cluster 4')
plt.scatter(X[pred == 4, 0], X[pred == 4, 1], s = 100, c = 'lightgreen', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('KMeans Cluster of Mall Customers')
plt.xlabel("Annual Income ($1000's)")
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
````

---

This appears to be a generally accurate KMeans model with an appropriate number of clusters. Some elements of Cluster 1 & 2, both on the right end of their clusters, could potentially be in their own cluster visually. 

Let's increase k to 7 and see what changes might be made to the data.

---

````python
kmeans = KMeans(n_clusters = 7, 
                init = 'k-means++',
                random_state = 20)
pred = kmeans.fit_predict(X)
````

````python
plt.figure(figsize=(12, 8))
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 100, c = 'darkred', label = 'Cluster 1')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'navy', label = 'Cluster 2')
plt.scatter(X[pred == 2, 0], X[pred == 2, 1], s = 100, c = 'forestgreen', label = 'Cluster 3')
plt.scatter(X[pred == 3, 0], X[pred == 3, 1], s = 100, c = 'darkcyan', label = 'Cluster 4')
plt.scatter(X[pred == 4, 0], X[pred == 4, 1], s = 100, c = 'lightgreen', label = 'Cluster 5')
plt.scatter(X[pred == 5, 0], X[pred == 5, 1], s = 100, c = 'mediumorchid', label = 'Cluster 6')
plt.scatter(X[pred == 6, 0], X[pred == 6, 1], s = 100, c = 'goldenrod', label = 'Cluster 7')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('KMeans Cluster of Mall Customers')
plt.xlabel("Annual Income ($1000's)")
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
````

---

Just as we predicted, the far-right sides of the original Cluster 1 & Cluster 2, now shown as Cluster 3 & Cluster 5, were split. However, there wasn't enough improvement in this KMeans model with an increase in cluster in our WCSS graph, so we'll stick to just 5 clusters for this dataset.
